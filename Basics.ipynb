```python
import csv
```

```python
f = open("guns.csv", "r")
csvreader = csv.reader(f)
data = list(csvreader)
data[:5]
```

```python
headers = data[0]
data.remove(headers)
print(headers)
print(data[:5])
```

```python
years = []
for row in data:
    years.append(row[1])
year_counts = {}
for year in years:
    if year not in year_counts:
        year_counts[year] = 1
    else:
        year_counts[year] = year_counts[year] + 1
        
print(year_counts)
```

```python
import datetime
dates = []
for row in data:
    dates.append(datetime.datetime(year = int(row[1]), month= int(row[2]), day=1))
dates[:5]
date_counts= {}
for date in dates:
    if date in date_counts:
        date_counts[date] = date_counts[date] + 1
    date_counts[date] = 0
date_counts
```
